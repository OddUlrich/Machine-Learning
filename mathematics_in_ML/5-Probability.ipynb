In order to avoid problems with division by zero, add one to the count of a value when computing the probability, and $1$ to the total, ensuring no probability will equal $0$. For example, if $m = 100$ and $1$ appeared $5$ times in a column $j$ of $Z$, then $pZ_j = \frac{1 + 5}{101}$.

```python
def compute_marginals(Z):
    pZ = np.zeros(Z.shape[1])
    
    #YOUR CODE HERE.
    m = Z.shape[0]
    pZ = (np.sum(Z, axis=0)+1) / (m+1)
    
    return pZ

print(compute_marginals(Y))
```

- the keys of $joint\_pZ$ are strings of $1s$ and $0s$ of length $n$, accounting for every possible length $n$ string of $1s$ and $0s$.
- the value associated with each key is the probability (between $0$ and $1$) of a row $z_{i}$ of $Z$ taking on the values represented by the key. This is equivalent to the joint probability of all of the elements of a row.
  
  
As with the computation of marginals, in order to avoid future problems with division by zero, add one to the count of a value when computing the probability, and $2^n$ to the total, ensuring no probability will equal $0$. For example, if $m = 100$, and $n=3$, and $111$ appeared $5$ times in $Z$, then $joint\_pZ[111] = \frac{1 + 5}{108}$.

Your function should work for any and all possible $m$ and $n$.

**HINT:** 
- To convert numbers to binary, look at how the dataset is generated.
- You might need to convert arrays of $1s$ and $0s$ to string form. $str(int(Z[ix, iy]))$ will work. 
- $joint\_pZ.keys()$ will return all the keys in the dictionary $joint\_pZ$.

---


```python
def compute_joints(Z):
    joint_pZ = {}
    n = Z.shape[1]
    m = Z.shape[0]
    cardinality = 2**n
    
    #YOUR CODE HERE.   
    
    # Build the base of dictionary
    for i in range(cardinality):
        joint_pZ[bin(i)[2:].zfill(n)] = 1
           
    # Count the distribution of each key.
    for i in range(m):
        key = ''
        for j in range(n):
            key += str(int(Z[i, j]))
        
        # Count in the dict.    
        joint_pZ[key] += 1
    
    for k, v in joint_pZ.items():
        v = v / (m+cardinality)    # The base of 1 has been added at the beginning.
        joint_pZ[k] = v

    return joint_pZ

print(compute_joints(Y)['0000'])
```

**TASK 1.3:** Complete the function $compute\_conditionals(Y, D) = conditional\_pY\_D$ that takes any matrices $D \in \{0, 1\}^{m \times n}$ and $Y \in \{0, 1\}^{m \times k}$ and returns a dictionary $conditional\_pY\_D$, where:
- the keys of $conditional\_pY\_D$ are strings of $1s$ and $0s$ of length $n$, accounting for every possible length $n$ string of $1s$ and $0s$.
- the value associated with each of those keys is another dictionary, the keys of which are every possible string of $1s$ and $0s$ of length $k$.
- the values associated with each of these length $k$ keys in the nested dictionaries is the conditional probability (between $0$ and $1$) of a row $y_{i}$ of $Y$ given the corresponding row $d_i$ of $D$. In other words, the conditional distribution $p(Y \mid X)$.
  
  
As with the computation of marginals and joints, in order to avoid future problems with division by zero, you must ensure no probability is $0$.

Your function should work for any and all possible $m$ and $n$.

```python
def compute_conditionals(Y, D):
    conditional_pY_D = {}
    
    #YOUR CODE HERE.
    m = D.shape[0]
    n = D.shape[1]
    k = Y.shape[1]

    # Joint event DY from the vertical conctenation between D and Y.
    DY = np.concatenate((D, Y), axis=1)
    
    joints_pD = compute_joints(D)    # p(D)
    joints_pY = compute_joints(Y)    # p(Y)
    joints_pDY = compute_joints(DY)  # p(D,Y)
    
    for d_key, d_value in joints_pD.items():
        conditional_pY_D[d_key] = {}
        
        for y_key, y_value in joints_pY.items():
            dy_key = d_key + y_key
            
            if dy_key in joints_pDY.keys():
                pY_D = joints_pDY[dy_key] / joints_pD[d_key] # p(Y|D) = p(Y,D) / p(D)
                conditional_pY_D[d_key][y_key] = pY_D

    return conditional_pY_D

#This code verifies you constructed the dictionary correctly.
conditional_Y_D = compute_conditionals(Y, D)
print('\nSample Test (should be near 0.36):')
print(conditional_Y_D['00010000']['0001'])
count = 0
verify = 0
for key in conditional_Y_D.keys():
    count += 1
    for subkey in conditional_Y_D[key].keys():
        verify += conditional_Y_D[key][subkey]
print('\nError is ', 100 * (verify / count - 1), '%.')
print('Floating point error around 2 % is acceptable and expected.')
```

First off, some definitions:

$X \in \mathbb{R}^{N \times m}$, where $N$ is the number of data points in your dataset $X$, and $m$ is the dimensionality of your data. For our example, $m=2$ and $N = 600$. $N$ is the number of rows, and $m$ is the number of columns of $X$.

$K$ is the total number of gaussians. This is just like the $K$ means you had for k-means in assignment 3. Each gaussian is named $k$, where $k$ is the unique number associated with that gaussian. For example, the first gaussian is $k = 1$. Each $k$ has a mean and covariance matrix associated with it. This allows you to construct a gaussian which is just a formula which can be used to generate data points or to compute the probability that a given data point is generated from this gaussian (look up generative models if you're curious to learn more). For this purpose, you can use np.random.multivariate_normal().

$\Sigma$ is a list of covariance matrices $\Sigma_k \in \mathbb{R}^{m \times m}$, which are symmetric positive semidefinite matrices.

$\mu$ is a list of means, each one $\mu_k \in \mathbb{R}^{m}$ associated with gaussian $k$.

$N$ is the total number of datapoints.

$responsibilities \in [0, 1]^{N \times K}$ is a matrix. Every column $k$ of $responsibilities$ is associated with the $k^{th}$ gaussian. Each element $r_{ik}$ of the $k^{th}$ column is the probability of the $i^{th}$ datapoint $x_i$ (the $i^{th}$ row of $X$) given the gaussian $k$.

$N_k$ is the sum of the $k^{th}$ column of $responsibilities$. In other words, there is one $N_k$ for each gaussian. $N_k = \sum_i r_{ik}$.

$\pi$ is a list of probabilies, 1 associated with each gaussian. $\pi_k$ is the probability of gaussian $k$. $\pi_k = \frac{N_k}{N}$.

   **TASK 2.1:** Complete the below function to initialise your parameters. You will need to generate:
   1. A list $\mu$ of $K$ means, each of which is in $\mathbb{R}^{m}$.
   2. A list $\Sigma$ of $K$ covariance matrices in $\mathbb{R}^{m \times m}$. Remember, covariance matrices must be symmetric positive semidefinite.
   3. A list $\pi$ of $K$ probabilities $\pi_k$. They should be initialised at $\frac{1}{K}$ (uniformly distributed at first).

Do not hard code parameters. You should generate them with some randomness, and your code must work for any $m$, $k$ and $N$.

You need to intelligently select $\mu_k$ just like you did with k-means. Poorly initialised parameters may result in an entirely broken EM algorithm.

```python
def initialise_parameters(X, K):
    m = X.shape[1]
    N = X.shape[0]
    sigma = []
    mu = []
    pi = []
    #YOUR CODE HERE
    
    ## Sigma initialization. ##
#     sigma_tmp = np.zeros((K, m, m))
#     for i in range(K):
#         mat = np.random.rand(m, m)
#         sigma_tmp[i] = mat@mat.T    # Symmetric and positive semidefinite.
#     sigma = sigma_tmp
    sigma_tmp = np.zeros((K, m, m))
    for i in range(K):
        sigma_tmp[i] = np.eye(m, m)
    sigma = sigma_tmp
    
    ## Mean initialization. ##
    
    # Randomly choose one point as the mean for the first gaussian.
    mu_tmp = np.zeros((K, m))
    mu_tmp[0] = X[np.random.randint(0, X.shape[0]), :]

    # Choose other mean for the rest gaussians.
    for i in range(1, K):
        dist = np.zeros(X.shape[0])
        p = np.zeros(X.shape[0])
        # Calculate the distance between x and chosen means.
        for j in range(X.shape[0]):
            dist[j] = np.sum(np.power((X[j] - mu_tmp[0]), 2))
            for k in range(1, i):
                # Save the distance between x and its nearest centroid.
                dist_tmp = np.sum(np.power((X[j] - mu_tmp[k]),2))
                if dist[j] != 0 and dist_tmp < dist[j]:
                    dist[j] = dist_tmp
        # Randomly choose a location of exactly one interval.
        p = dist / np.sum(dist)
        loc = np.random.uniform(0, 1)
        p_sum = 0
        for idx in range(len(p)):
            if loc > p_sum and loc < p_sum+p[idx]:
                break;
            else:
                p_sum += p[idx]
        mu_tmp[i] = X[idx, :]
        
    mu = mu_tmp
    
    ## Pi initialization ##
    for k in range(K):
        pi.append(1/K)
    
    return sigma, mu, pi
    
K = 3
sigma, mu, pi = initialise_parameters(X, K)
print('\nSigma: \n', sigma)
print('\nMu: \n', mu)
print('\nPi: \n', pi)
```

$E\_step$ computes the matrix $responsibilities \in [0, 1]^{N \times K}$, where $N$ is the number of data points, and $K$ is the number of gaussians you're attempting to cluster the data with. Each gaussian will be associated with a column of $responsibilities$. As your algorithm runs, each row represents a data point $x_i$, and each column of that row will contain the probability that that $x_i$ came from that gaussian, $p(x_i \mid k)$, signifying the extent to which this datapoint $x_i$ has been assigned to the gaussian associated with that column. 

---

**TASK 2.2:** Implement $E\_step(\pi, \mu, \Sigma, X) = responsibilities$, which updates the $responsibilities$ matrix.

Remember, which gaussian a datapoint is assigned to depends on the probability of that datapoint given that guassian.

You cannot use euclidean distance like you did with the k-means algorithm in assignment 3.

**HINT:** 
- Each element of responsibilities $r_{ik} = \frac{\pi_k \mathcal{N}(x_i \mid \mu_k \Sigma_k)}{\sum_j \pi_j \mathcal{N}(x_i \mid \mu_j \Sigma_j)}$, where $x_i$ is the $i^{th}$ row of $X$.
- https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.stats.multivariate_normal.html
- Use $multivariate\_normal.pdf(x, \mu_k, \Sigma_k)$

---

```python
def E_step(pi, mu, sigma, X):
    N = X.shape[0]
    m = X.shape[1]
    K = len(pi)
    responsibilities = np.zeros((N, K))
    #YOUR CODE HERE
    epsilon = 1.0e-9
    
    # Calculae the weighted distribution w.r.t every date point in different gaussian distribution.
    for i in range(N):
        for k in range(K):
            responsibilities[i, k] = pi[k] * multivariate_normal.pdf(X[i], mean=mu[k], cov=sigma[k]) + epsilon
        
    # Divided by the sum of weighted distribution for each data point w.r.t every gaussian distribution.
    for i in range(N):
        responsibilities[i, :] /= np.sum(responsibilities[i])
    return responsibilities

responsibilities = E_step(pi, mu, sigma, X)
print(responsibilities)
```

**TASK 2.3:** Implement $M\_step(responsibilities, X) = \mu, \Sigma, \pi$ which returns the updated means and covariances for all of the $k$ gaussians, along with the priors $\pi$.

**HINT:** 
- https://docs.scipy.org/doc/numpy/reference/generated/numpy.outer.html

---

```python
def M_step(responsibilities, X):
    K = responsibilities.shape[1]
    N = X.shape[0]
    m = X.shape[1]
    mu = []
    sigma = []
    pi = []
    
    #YOUR CODE HERE
    n_k_list = np.sum(responsibilities, axis=0)
    mu_tmp = np.zeros((K, m))
    sigma_tmp = np.zeros((K, m, m))
    
    for k in range(K):
        # Update mu.
        weighted_X = np.zeros_like(X)
        for idx in range(m):
            weighted_X[:, idx] = responsibilities[:, k] * X[:, idx]
        
        mu_tmp[k] = np.sum(weighted_X, axis=0) / n_k_list[k]
        
        # Update sigma.
        var = np.zeros((N, m, m))
        for i in range(N):
            var[i] = responsibilities[i, k]*(np.asmatrix(X[i] - mu_tmp[k]).T@np.asmatrix(X[i] - mu_tmp[k]))
        sigma_tmp[k] = np.sum(var, axis=0) / n_k_list[k]
        
        
    mu = mu_tmp
    sigma = sigma_tmp
    pi = n_k_list / N
    
    return mu, sigma, pi

mu, sigma, pi = M_step(responsibilities, X)
print('\nSigma: \n', sigma)
print('\nMu: \n', mu)
print('\nPi: \n', pi)
```

**TASK 2.4:** Implement $classify(\pi, \mu, \Sigma, x) = k$ which takes an unknown example $x \in \mathbb{R}^{m}$, the means $\mu$ and covariance matrices $\Sigma$ and the priors $\pi$, and returns $k$, the number of the gaussian which maximises the probability of $x$.

In other words, ${arg max}_k \left(p(x \mid \mu_k, \Sigma_k)\pi_k \right)$.

---

```python
def classify(pi, mu, sigma, x):
    k = None
    K = len(pi)
    #YOUR CODE HERE
    p = np.zeros(K)
    
    for i in range(K):
        p[i] = pi[i] * multivariate_normal.pdf(x, mean=mu[i], cov=sigma[i]) 
    
    k = np.argmax(p)
    return k

print(classify(pi, mu, sigma, X[0, :]))
```

```python
def EM(X, K, iterations):
    N = X.shape[0]
    #YOUR CODE HERE
    
    sigma, mu, pi = initialise_parameters(X, K)
    for i in range(iterations):
        responsibilities = E_step(pi, mu, sigma, X)
        mu, sigma, pi = M_step(responsibilities, X)
    
    return mu, sigma, pi #mu, sigma, pi


#Test code. Leave it aloooooone!
iterations = 10
K = 3
mu, sigma, pi = EM(X, K, iterations)
print('\nSigma: \n', sigma)
print('\nMu: \n', mu)
print('\nPi: \n', pi)

def allocator(pi, mu, sigma, X, k):
    N = X.shape[0]
    cluster = []
    for ix in range(N):
        prospective_k = classify(pi, mu, sigma, X[ix, :])
        if prospective_k == k:
            cluster.append(X[ix, :])
    return np.asarray(cluster)

colours = ['r', 'g', 'b']
for k in range(K):
    cluster = allocator(pi, mu, sigma, X, k)
    plt.scatter(cluster[:,0], cluster[:,1], c=colours[k])
plt.show()
```
