   **TASK 1.1:** Complete the function $laplace\_expansion(A) = |A|$, where $A \in \mathbb{R}^{D \times N}$, and $|A|$ is the determinant of $A$.
   
**HINT:**
- One way to find the determinant is to implement the Laplace Expansion algorithm. This is described on slide 4 of the matrix decomposition lecture slides.
- $|A| \neq 0$ iff $A$ is invertible.
- Not all matrices have a determinant. If a matrix doesn't have a determinant, return 'False' or some other indication that no determinant was computable.

---


```python
def det(A):
    det_A = None
    #YOUR CODE HERE
    
    m = A.shape[0]
    n = A.shape[1]
    
    if m != n:
        print("Input matrix is invertible!")
        return False
    
    # A.shape = (1, 1)
    if m == 1 and n == 1:
        return A[0][0]
    
    det_A = 0
    
    # Laplace expansion along the first column.
    for row in range(m):  # 0 ~ m-1
        # Submatrix of A.
        sub_A = np.zeros((m-1, n-1))
        
        if row == 0:
            sub_A = A[1:, 1:]
        elif row == (m-1):
            sub_A = A[0:m-1, 1:]
        else:    
            sub_A[:row, :] = A[0:row, 1:]
            sub_A[row:, :] = A[row+1:, 1:]
        
        if row % 2 == 0:  # plus 
            det_A += A[row][0] * det(sub_A)
        else:  # substraction
            det_A -= A[row][0] * det(sub_A)
        
    return det_A

print(det(np.eye(2)))
print(det(np.zeros((2, 3))))
```

**TASK 1.3:** Let $A\_{pp} \in \mathbb{R}^{D \times N}$ be a matrix of data. Each column of $A\_{pp}$ is a sample of data (1 training example for instance). The rows of $A\_{pp}$ are thus the features (dimensions) of each of these samples. Complete the function $preprocess(A\_{pp}) = A, Q\_norms, A\_means$, for which:

$$Q_{i,:} = A\_{pp}_{i,:} - \mu_i$$

...where $\mu_i = \frac{1}{m}\sum_j A\_{pp}_{ij}$ or $\mu_i = \frac{min(A\_{pp}_{i,:}) + max(A\_{pp}_{i,:})}{2}$. You can choose which, it won't affect your grade (if you're wondering why, think about the theory assignment solutions you saw for A4).

$$A_{i,:} = \frac{Q_{i,:}}{||Q_{i,:}||_\infty }$$


$A \in \mathbb{R}^{D \times N}$

$Q_{i,:}$ is the $i^{th}$ row of $Q$.

$A_{i,:}$ is the $i^{th}$ row of $A$.

$||Q_{i,:}||_\infty$ is the infinity norm of $Q_{i,:}$.

$Q\_norms \in \mathbb{R}^{D}$ is a vector recording $||Q_{i,:}||_\infty$ for every feature dimension $i$.

$A\_means \in \mathbb{R}^{D}$ is a vector recording $\mu_i$ for every feature dimension $i$.


**HINT:** 
- np.amax and np.abs may be useful.
- If the norm is 0, divide by 1 instead.

---

```python
def preprocess(A_pp):
    A = None
    Q_norms = None
    A_means = None
    #YOUR CODE HERE
    
    D = A_pp.shape[0]
    N = A_pp.shape[1]
    
    # A_means (D, 1)
#     A_means = np.sum(A_pp, axis=1) / N
    A_means = (np.min(A_pp) + np.max(A_pp)) / 2
    
    # Q_norms = ||Q_i:||_infty = max|Q_i| (D, 1)
    Q = np.zeros((D, N))
    A_means_tile = np.tile(A_means, (N, 1)).T    # (D, 1) --> (N, D) --> (D, N)
    Q = A_pp - A_means_tile
    
    Q_norms = np.amax(np.abs(Q), axis=1)
    Q_norms[np.argwhere(Q_norms == 0)] = 1    
    Q_norms_tile = np.tile(Q_norms, (N, 1)).T    # (D, 1) --> (N, D) --> (D, N)  
    
    # A (D, N)
    A = np.zeros((D, N))
    A = Q / Q_norms_tile
    
    return A, Q_norms, A_means

A, Q_norms, A_means = preprocess(A_pp)
print(A.shape)
print(Q_norms)
print(A_means)
```

The above function implements the standardization step for each feature dimension in all data samples.

The variances(or the ranges) in different dimensions can be various, and the standardization not only finds out the all variances (eigenvalues) in all dimensions, but also scales all the data values into a range from -1 to 1, which can reduce the influences from small changes in a dimension with low variance, and reserve the effects from small changes in a dimension with high variance due to the floating point underflow. It can reduce the information loss as much as possible.

Infinity norm is sensitive to the outlier and this property satisfies the need to scale down data into the range from -1 to 1, while L2 norm is more likely to consider the distribution of all the data point, which cannot implement the standardization step.

Our objective is to describe each image of a face as a linear combination of other images. These other images are called Eigen Faces. The Eigen Faces, when flattenned into a column vector, is $77760\times 1$ dimensional. These vectors are the Eigenvectors of the matrix $AA^T$.

To compute the eigen faces $F$, we need to compute the eigenvalues and eigenvectors of matrix $A^T A $. From our lecture, you should be aware that the non-zero eigenvalues of $A A^T$ and $A^T A$ are identical. 

Moreover, if we let one of the eigenvectors of $A^T A $ be $v_m$, then the corresponding eigenvector of $AA^T$ is $f_m=A v_m$. 


So...

$$F = [f_1, f_2, ..., f_N]\in \mathbb{R}^{77760 \times 165}$$

To represent our 165 images as a linear combination of the 165 principle components (eigenfaces), we project our images onto the 165 principle components, and obtain

$$C = A^T F\in \mathbb{R}^{165 \times 165}$$.

```python
def eigen_faces(A_pp):
    C = None
    F = None
    D = None
    Q_norms = None
    A_means = None
    #YOUR CODE HERE
    
    # A_pp:(D, N), A:(D, N), Q_norms:(D, 1), A_means:(D, N)
    A, Q_norms, A_means = preprocess(A_pp)

    # D: eigenvalue matrix
    eig_val, eig_vec = np.linalg.eig(A.T@A)    # A.T@A: (N, N), eig_val: (N, 1), eig_vec:(N, N)
    D = np.diag(eig_val.real)

    # F: (D, N)
    F = A@eig_vec.real
    F_norm = np.linalg.norm(F, axis=0)
    F_norm_tile = np.tile(F_norm.T, (F.shape[0], 1))
    F /= F_norm_tile
    
    # C: (N, N)
    C = A.T@F    
    
    return C, F, D, Q_norms, A_means

#For the purposes of doing this assignment, this code isn't really here. Pretend it's engraved in rock.
C, F, D, Q_norms, A_means = eigen_faces(A_pp)
print('Orthogonality Check (should be close to 0): ', F[:, 0].T@F[:, 1])
print('Unit Vector Check: ', math.isclose(F[:, 0].T@F[:, 0], 1))
print(C.shape)
print(F.shape)
```

Recall assignment 2, in which you rotated a vector until you found the largest value for the projection of the data onto that vector.

This was the eigen vector corresponding to the largest eigen value of that dataset. It was the "direction of greatest variance".

An eigen value $d_j$ in the $j^{th}$ column of $D$ (the matrix of eigen values), is associated with an eigen vector (or eigen face) in the $j^{th}$ column $f_j$ of $F$.

The variance of the data in the direction of $f_j$ is exactly $d_j$. So the $\sum_j d_j$ is the total variance of the dataset in all directions.

When we project the dataset onto an eigen face $f_k$, we "capture" a percentage of this variance $\frac{d_k}{\sum_j d_j}$. For example, if $\sum_j d_j = 100$, and we project our data on to the 2 eigen faces $f_1$ and $f_2$ with corresponding eigen values $d_1 = 19$ and $d_2 = 3$, then we've captured $\frac{22}{100} = 22\%$ of the variance of the dataset.

```python
def reduce_dimensionality(image_vector, k, F, D, A_means, Q_norms):
    compressed_image = None
    p = None
    #YOUR CODE HERE
    
    # Preprocess the image vector.
    img_normed_vec = (image_vector - A_means) / Q_norms
    
    # Extract the eigenvalue from matrix D to calculate p.
    one_vec = np.ones(D.shape[0])
    eig_val = D@one_vec
    p = np.sum(eig_val[:k]) / np.sum(eig_val)
    
    # C_i @ F: compressed_image is the combination of the first k^th eigen images.
    F_k = np.zeros_like(F)
    F_k[:, :k] = F[:, :k]
    compressed_image = img_normed_vec.T@F_k
    
    return compressed_image, p


#Display Code. Leave it alooooooooooone.
#You can mess with settings, but return them to their original values.
compressed_image, p = reduce_dimensionality(A_pp[:, Idx], 10, F, D, A_means, Q_norms)

print('Variance Captured:', int(p * 100), '%')

R_c = reconstruct_image(compressed_image, F, Q_norms, A_means)
plt.imshow(R_c)
print(np.amax(R_c), np.amin(R_c))
plt.show()
R_o = reconstruct_image(Img, F, Q_norms, A_means)
plt.imshow(R_o)
print(np.amax(R_o), np.amin(R_o))
plt.show()
print('Error (expect around 1700000 for k = 10, 0 for k = 165): \n', np.sum(np.abs(R_c - R_o)))
```
