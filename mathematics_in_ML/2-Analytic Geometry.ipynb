Define any $ x = [x_1, x_2]^T \in R^2, y = [y_1, y_2]^T \in R^2, z = [z_1, z_2]^T \in R^2 $

**1) Bilinear mapping**

Define $\lambda, \varphi \in R$ , then
$$
\begin{align}
\langle \lambda x + \varphi y , z \rangle & = (\lambda x_1 + \varphi y_1)z_1 - [(\lambda x_1 + \varphi y_1)z_2 + (\lambda x_2 + \varphi y_2)z_1] + 2(\lambda x_2 + \varphi y_2)z_2 \\
&= \lambda[x_1z_1 - (x_1z_2 + x_2z_1) + 2x_2z_2] + \varphi[y_1z_1 - (y_1z_2 + y_2z_1) + 2y_2z_2] \\
&= \lambda \langle x, z \rangle + \varphi \langle y, z \rangle
\end{align}
$$

Similarly, 
$$
\begin{align}
\langle  x, \lambda y + \varphi z \rangle &= \lambda \langle x, y \rangle + \varphi \langle x, z \rangle
\end{align}
$$

**2) Symmetric**
$$
\begin{align}
\langle x, y \rangle &= x_1y_1 - (x_1y_2 + x_2y_1) + 2x_2y_1 \\
\langle y, x \rangle &= y_1x_1 - (y_1x_2 + y_2x_1) + 2y_2x_1 \\
\langle x, y \rangle &= \langle y, x \rangle
\end{align}
$$

**3) Positive definite**

For $ x = 0 $,
$$
\begin{align}
\langle x, x \rangle &= \langle 0, 0 \rangle = 0 
\end{align}
$$

For $ \forall x \in R^2\setminus\{0\}$ ,
$$
\begin{align}
\langle x, x \rangle &= x_1^2 - (x_1x_2 + x_2x_1) +2x_2^2 \\
&= (x_1 - x_2)^2 + x_2^2 \gt 0
\end{align}
$$

So from the analysis above, the $\langle \cdot, \cdot \rangle$ defined is an inner product.

---
   **TASK 1.1:** Create a function $find\_angle(a, b)$ that:
   - Takes as input 2 parameters:
       - a 3x1 matrix $a$ (a 3 dimensional vector).
       - a 3x1 matrix $b$ (a 3 dimensional vector).
   - Returns $d$, the angle (in degrees) between the 2 vectors.

In other words, compute the angle between 2 lines originating at the origin.

---

**HINT:**
- https://onlinemschool.com/math/library/vector/angl/
- https://docs.scipy.org/doc/numpy/reference/generated/numpy.linalg.norm.html
- https://docs.scipy.org/doc/numpy/reference/generated/numpy.sqrt.html
- https://docs.scipy.org/doc/numpy/reference/generated/numpy.arccos.html
- https://docs.scipy.org/doc/numpy/reference/generated/numpy.degrees.html

```python
a = np.array([[-0.577],
              [0.577],
              [0.577]])

b = np.array([[1],
              [0],
              [0]])

def find_angle(a, b):
    #YOUR CODE HERE
    # Use dot product as inner product
    product = np.dot(a.T, b)
    norm_a = np.sqrt(np.dot(a.T, a))
    norm_b = np.sqrt(np.dot(b.T, b))
#     norm_a = np.linalg.norm(a)
#     norm_b = np.linalg.norm(b)
    
    val = product / norm_a / norm_b
    rad = np.arccos(val)  # 0 - PI
    
    d = rad * 180 / np.pi
    d = np.nan_to_num(d)

    
    return d[0][0] #d[0][0]

print(find_angle(a, b))
```

**TASK 1.2:** Define a function $rotator(v, \theta_x, \theta_y, \theta_z)$ that takes a 3 dimensional vector $v \in \mathbb{R}^{3 \times 1}$ (or a matrix in $\mathbb{R}^{3 \times N}$ if you want), and rotates it $\theta_x$ degrees around the $x$ axis, $\theta_y$ degrees around the $y$ axis, and $\theta_z$ degrees around the $z$ axis.

---

**HINT:**
- You need to generate 3 rotation matrices and apply each of them to $v$.
- https://en.wikipedia.org/wiki/Rotation_matrix#In_three_dimensions
- Use $math.radians()$ to convert degrees to radians.
- Use $math.cos()$ and $math.sin$, but remember those functions take radians, not degrees!
- There may be some minor numerical errors. Don't worry, this isn't a floating point algebra course, you won't be penalised for floating point error! If you don't understand what I just said, you also won't be penalised!

```python
def rotator(v, theta_x, theta_y, theta_z):
    #YOUR CODE HERE
    
    x_rad = math.radians(theta_x)
    y_rad = math.radians(theta_y)
    z_rad = math.radians(theta_z)
    
    Rx = np.array([[1, 0, 0],
                   [0, math.cos(x_rad), -math.sin(x_rad)],
                   [0, math.sin(x_rad), math.cos(x_rad)]])
    
    Ry = np.array([[math.cos(y_rad), 0, math.sin(y_rad)],
                   [0, 1, 0],
                   [-math.sin(y_rad), 0, math.cos(y_rad)]])
    
    Rz = np.array([[math.cos(z_rad), -math.sin(z_rad), 0],
                   [math.sin(z_rad), math.cos(z_rad), 0],
                   [0, 0, 1]])
    
    return Rx@Ry@Rz@v

v = np.array([[1],
              [0],
              [0]])

print(rotator(v, 0, 0, 90))
```

The projection of a vector $a$ onto another vector $b$ *sounds* intimidating, but really it's just the amount to which $a$ and $b$ point in the same direction, multiplied by the direction $b$. 

---
   **TASK 2.1:** Create a function $projector(a, b) = c$ that takes as input $a$ and $b$, and returns the projection of $a$ onto $b$.

---

**HINT:** https://en.wikipedia.org/wiki/Vector_projection

```python
def projector(a, b):
    #YOUR CODE HERE.
    
    # 1. Find the coordinate 𝜆.
    lamb = np.dot(b.T, a) / np.dot(b.T, b)
    
    # 2. Find the projection point 𝜋.
    c = lamb * b
    
    return c

a = np.array([[-0.577],
              [0.577],
              [0.577]])

b = np.array([[1],
              [0],
              [0.5]])

print(projector(a, b))
```

You now have $v$ which maximises $avg\_projection\_length(v, X)$!

Fun fact which may be incomprehensible now but will be really important once you get the hang of machine learning; this $v$ is the direction of largest variance for $X$. That means it is the eigen vector that is associated with the largest eigen value (in this case singular value, because $X$ has no inverse) of $X$!

```python
transform = np.array([[0.05, 0, 0],
                     [0, 0.65, 0.5],
                     [0.1, -0.3, 0.35]])
X = np.random.randn(100, 3) @ transform
v = np.array([[1], [0], [0]])


def avg_projection_length(v, X):
    n, m = X.shape #n is the number of rows X has. m is the number of columns.
    p = 0
    #YOUR CODE HERE
    
    for i in range(n):
        angle = find_angle(X[i, :].T, v)
        p += abs(math.cos(math.radians(angle))) * np.linalg.norm(X[i, :])
        
    return p / n

def find_data_direction(v, X):
    #YOUR CODE HERE
    
    mark_v = v
    max_len = avg_projection_length(v, X)
    
    for theta_x in range(0, 360, 30):
        for theta_y in range(0, 360, 30):
            for theta_z in range(0, 360, 30):
                new_v = rotator(v, theta_x, theta_y, theta_z)
                tmp_len = avg_projection_length(new_v, X)
                if tmp_len > max_len:
                    max_len = tmp_len
                    mark_v = new_v
    v = mark_v
    
    return v


#PLOTTING CODE - IGNORE! OR AT LEAST DON'T MESS WITH IT!
print('\nAverage Projection Length:')
print(avg_projection_length(v, X))
new_v = find_data_direction(v, X)
print('\nAverage Projection Length After Rotation:')
print(avg_projection_length(new_v, X))
origin = [0], [0], [0]
placeholder = np.array([[[0], [0], [0]]])
fig = plt.figure(figsize=(16, 10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:, 0], X[:, 1], X[:, 2], alpha=0.45)
ax.quiver(*origin, new_v[0, :], new_v[1, :], new_v[2, :], color = 'red')
ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)
ax.set_zlim(-1.5, 1.5)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```

So, now (hopefully) you have a vector that points in the "same direction" as the data. 

In every direction, the data has variance. In some directions, the variance is very small, in others very large.

But what does this mean? Well, currently we're describing the data in 3 dimensions described by the following 3 columns:

$$\begin{bmatrix}
1&0&0\\
0&1&0\\
0&0&1\\
\end{bmatrix}
$$

Each column is a vector. The 1 in each column represents a step of one along the dimension represented by that row.

For data we collect, each "feature" is a dimension.

But we can describe data more efficiently, compressing the data to make it easier for algorithms to learn.

This is what we took the first step towards above, when we found the direction of greatest variance for the dataset!

---

**TASK 3.1:** Project the dataset $X$ onto your updated version of $v$, the direction of greatest variance, to get $X_c$, the compressed version of $X$.

---

**HINT:** 
- Remember $X$ is a 100 by 3 matrix, and $v$ is a 3 dimensional vector. How do you project 100 examples onto a 3 dimensional vector? 
- Remember, $v$ is a unit vector. It should already have a length of 1.

```python
v = new_v
X_c = np.zeros_like(X)
#YOUR CODE HERE
for i in range(X.shape[0]):
    X_c[i, :] = projector(X[i, :], v).T
```

Great. You now have a 100 dimensional vector of coefficients. I bet you know exactly how useful this is, but in case you don't here's rundown.

Each number in $X_c$ represents a datapoint. When you multiply that number by $v$, you get the 3 dimensional representation of that datapoint back (but only in terms of $v$). 

In other words, each number is a compressed version of a 3 dimensional point. To uncompress it, multiply each number by $v$ again!

---
**TASK 3.2:** Create a 100 by 3 vector, each row of which is $v^T$ ($v$ transposed). Then piecewise multiply this vector using * with $X_c$ to get the uncompressed version of  $X_c$! 

---

**HINT:**
- You can stack 100 copies of $v^T$ using the tile function: https://docs.scipy.org/doc/numpy/reference/generated/numpy.tile.html
- Piecewise multiplication of vectors is done using *. $v_1*v_2$ gives you element 1 of $v_1$ times element 1 of $v_2$, element 2 of $v_1$ times element 2 of $v_2$ etc...
- If you're stuck, you can use a loop instead of stacking $v$ to do this with tile.

**NOTE:** X_uncompressed isn't quite the same as $X$. This is because what we've done is an example of lossy compression! This lossy compression saved $\frac{2}{3}$ of the space, but more importantly any algorithm has to only learn from data $\frac{1}{3}$ as complex (though complexity is much more complex than this in reality). That's enough to make many problems salvageable and manageable! As you'll learn later, "dimensionality reduction" as this is called is a critical part of machine learning. 

```python
#YOUR CODE HERE
v_stack = np.tile(v.T, (X_c.shape[0], 1))
X_uncompressed = X_c*v_stack


#PLOTTING CODE - IGNORE!
origin = [0], [0], [0]
placeholder = np.array([[[0], [0], [0]]])
fig = plt.figure(figsize=(16, 10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_uncompressed[:, 0], X_uncompressed[:, 1], X_uncompressed[:, 2], alpha=0.45)
ax.scatter(X[:, 0], X[:, 1], X[:, 2], alpha=0.15, color='red')
ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)
ax.set_zlim(-1.5, 1.5)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
print('RED = original X')
print('BLUE = your compressed and uncompressed X')
plt.show()
```
